```python
mydf=pd.read_sql("select sum(i.sellingprice*s.quantity) as yearly_sales,extract(year from s.created_at) as year,sum(s.quantity) as total_quantity,sum(st.stock) total_stock,count(s.id) total_products_sold,round(avg(i.buyingprice),2)buying_price,round(avg(i.sellingprice),2)selling_price from sales s join inventories as i on s.inv_id = i.id join stock as st on st.inv_id=i.id group by year order BY year asc",conn)
```

```python
y=mydf[['yearly_sales']]
```
